<h2 id="identificar_lidar_valores_faltantes">Identificar e Lidar com valores faltantes</h2>


<h3 id="identificar_valores_faltantes">Identificar valores faltantes</h3>
<h4>Convertendo "?" para NaN</h4>
No conjunto de dados de veículos, dados faltantes aparecem como pontos de interrogação "?". É possível substituir os "?" por "NaN" (Not a Number), que é o marcador do Python para valores faltantes. Por razões de velocidade computacional e conveniência. Para fazer esta troca iremos utilizar o comando <code>dataframe.replace(A, B, inplace = True)</code> para trocar A por B.  

Note que o "NaN" é um valor que está na biblioteca <b>numpy</b> do Python, por isso precisamos importar esta biblioteca, conforme apresentado abaixo:

```python
import numpy as np
frame_dados.replace("?", np.nan, inplace = True)
frame_dados.head()
```

<h4>Verificando valores faltantes</h4>

Os dados faltantes foram convertidos para o padrão utilizado pelo Python. Iremos utilizar alguns métodos do Python para identificar os valores faltantes. Existem dois métodos para identificar dados faltantes: 
<ol>
    <li><b>.isnull()</b></li>
    <li><b>.notnull()</b></li>
</ol>
A saída destes métodos é um valor booleano indicando se o valor que foi passado como argumento é ou não um dado faltante. Veja no exemplo mostrado abaixo: 

```python
dados_faltantes = frame_dados.isnull()
dados_faltantes.head(5)
```

Neste caso, **True** significa que o dado está faltando no frame de dados e **False** significa que o dado existe no frame de dados.

<h4>Contando os dados faltantes em cada coluna</h4>
<p>
Utilizando um loop do Python podemos rapidamente verificar o número de dados faltantes em cada coluna do frame de dados. Como mencionado acima <b>True</b> representa um dado faltante e <b>False</b> significa um valor no frame de dados, portanto, no corpo do loop o método <code> .value_counts()</code>  contará o número de valores que são <b>True</b>.
  
</p>

```python
for coluna in dados_faltantes.columns.values.tolist():
    print(coluna)
    print (dados_faltantes[coluna].value_counts())
    print("")    
```

Baseado no resultado apresentado acima, cada coluna possui 205 linhas de dados e 7 colunas apresentam dados faltantes:

    <li>"tempos" : 4 dados faltantes</li>
    <li>"potência(HP)": 2 dados faltantes</li>
    <li>"rpm-máximo": 2 dados faltantes</li>
    <li>"preco": 4 dados faltantes</li>

<h3 id="lidando_com_dados_faltantes">Lidando com dados faltantes</h3>
<b>Como lidar com dados faltantes?</b>

<ol>
    <li>elimine os dados<br>
        a. elimine toda a linha<br>
        b. elimine toda a coluna
    </li>
    <li>troque os dados<br>
        a. substitua pela média<br>
        b. substitua pela frequência<br>
        c. substitua baseado em alguma outra função.
    </li>
</ol>

Uma coluna deve ser eliminada somente se a maioria das entradas na coluna estiverem vazias. Em nosso caso nenhuma coluna possui valores faltantes que justifique eliminar a coluna como um todo. 
Nós temos alguma liberdade para escolher o melhor método para substituir valores faltantes, porém, alguns métodos podem ser mais razoáveis que outros. Vamos aplicar alguns métodos para colunas diferentes e verificar o que ocorre:

<b>Substituir pela média:</b>
<ul>
    <li>"perda-normalizada": 41 dados faltantes, substituir os dados pela média</li>
    <li>"tempos": 4 dados faltantes, substituir os dados pela média</li>
    <li>"diâmetro-cilindro": 4 dados faltantes, substituir os dados pela média</li>
    <li>"potência(HP)": 2 dados faltantes, substituir os dados pela média</li>
    <li>"rpm-máximo": 2 dados faltantes, substituir os dados pela média</li>
</ul>

<b>Substituir pela frequência:</b>
<ul>
    <li>"num-de-portas": 2 dados faltantes, substituir os dados por "four". 
        <ul>
            <li>Motivo: 84% dos sedans possuem 4 portas. Como 4 portas é o mais frequente este valor é o mais provável de ocorrer.</li>
        </ul>
    </li>
</ul>

<b>Eliminar a linha toda:</b>
<ul>
    <li>"preco": 4 dados faltantes, simplesmente eliminamos a linha toda.
        <ul>
            <li>Motivo: preço é o que queremos predizer. Cada entrada de dados sem preço não pode ser utilizada para predição, portanto a linha não é útil.  </li>
        </ul>
    </li>
</ul>

<p> Vamos na sequência determinar a média das colunas "perda-normalizada" e "diâmetro-cilindro" e substituir os valores "NaN" pela média encontrada.</p>

```python
## Determina as médias de cada coluna e imprime os valores encontrados

media_perda_norm = frame_dados["perda-normalizada"].astype("float").mean(axis=0)
print("Média de perda-normalizada:", media_perda_norm)
media_diam_cilindro = frame_dados["diâmetro-cilindro"].astype('float').mean(axis=0)
print("Média de diâmetro de cilindro:", media_diam_cilindro)
```

```python
## Substitui os valores NaN pelas médias encontradas
frame_dados["perda-normalizada"].replace(np.nan, media_perda_norm, inplace=True)
frame_dados["diâmetro-cilindro"].replace(np.nan, media_diam_cilindro, inplace=True)
```

### Exercício 4: ###

Determine a média e faça a substituição dos valores NaN pela média para as demais colunas do frame de dados.

```python
## Escreva os códigos abaixo calculando as médias primeiramente e depois fazendo as substituições 
media_tempos = frame_dados['tempos'].astype("float").mean(axis = 0)
media_potencia_hp = frame_dados['potência(HP)'].astype("float").mean(axis = 0)
media_rpm_max = frame_dados['rpm-máximo'].astype("float").mean(axis = 0)

print("Média tempos: ", media_tempos)
print("Media potência: ", media_potencia_hp)
print("Média RPM max: ", media_rpm_max)

frame_dados["tempos"].replace(np.nan, media_tempos, inplace = True)
frame_dados["potência(HP)"].replace(np.nan, media_potencia_hp, inplace = True)
frame_dados["rpm-máximo"].replace(np.nan, media_rpm_max, inplace = True)
```

<h2 id="padronizacao_dados">Padronização de Dados</h2>
<p>
Os dados muitas vezes são obtidos de fontes diferentes e em formatos diferentes. A padronização dos dados é um termo utilizado para um tipo específico de normalização de dados, onde subtraímos dos dados a sua média e dividimos os valores obtidos pelo desvio padrão. 
</p>
    
<b>O que é padronização?</b>
<p>Padronização é um processo de transformação de dados em um formato comum que permite ao pesquisador fazer comparações mais significativas.  
</p>

<b>Exemplo</b>
<p>Transformar mpg para km/l:</p>
<p>Em nosso frame de dados o consumo de combustível foi obtido em mpg para a cidade e para a estrada. Aqui no Brasil esta unidade de medida não é significativa e a maioria não é capaz de identificar se o carro é ou não economico.</p>
<p>Vamos então converter, aplicando uma <b>transformaçào de dados</b> para transformar mpg em km/l.</p>

<p>A fórmula para converter a unidade é dada abaixo:<p>
    1 milha = 1.60934 km<p>
    1 galão = 3.78541 l<p>
    logo: mpg = 1.60934/3.78541 km/l, ou mpg = 0.4251429 km/l 
<p>Várias operações matemáticas podem ser escritas direto no Pandas.</p>



```python
## Converte mpg para km/l conforme indicado acima.
frame_dados['consumo-cidade(km/l)'] = 0.4251429*frame_dados['consumo-cidade(mpg)']

## Verificando a transformação
frame_dados.head()
```

<h2 id="histogramas">Histogramas</h2>
<b>Porque histogramas?</b>
<p>
    Um histograma é um processo de transformar variáveis numéricas e contínuas em classes discretas de valores, facilitando uma análise de agrupamentos. 
</p>

<b>Exemplo: </b>
<p>Em nosso frame de dados a coluna de "potência(HP)" é uma variável inteira que varia entre 48 e 262 HPs com 59 valores únicos. Vamos assumir que estamos interessados em diferentes preços de carros com alta potência, média potência e baixa potência (3 tipos). Seria possível rearranjar os dados em 3 classes para simplificar a análise?  </p>

<p>Vamos utilizar o método <code>.cut</code> para segmentar a coluna de "potência(HP)" em 3 classes.  </p>

<h3>Exemplo de segmentação de dados em Panda</h3>

 Vamos utilizar a biblioteca <b> matplotlib</b> para plotar o histograma de potência para ver a distribição como a distribuição de potência se apresenta.
 

```python
## Carrega a biblioteca matplotlib e as funções necessárias para criar o gráfico
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

## Cria um histograma da coluna de potência
plt.pyplot.hist(frame_dados["potência(HP)"])

## Ajusta os valores dos eixos e do título do gráfico
plt.pyplot.xlabel("Pontência(HP)")
plt.pyplot.ylabel("quantidade")
plt.pyplot.title("classes de potência")
```

<p>Note que o histograma apresentado agrupou as potências em 10 grupos distintos. Como mencionado acima o desejo era de ter apenas 3 classes no histograma, para isto temos que utilizar a biblioteca <b>numpy</b> com o método <code>linspace(valor_inicial, valor_final, numero_valores)</code>. Note que para dividirmos as potências em 3 classes precisamos de 4 valores para separar os dados. </p>

A lista de valors gerados acima será utilizada para criar as classes de potência usando o método <code>.cut(coluna, valores, rótulos, tags)</code> 

Na sequência definimos os nomes das classes a serem utilizadas na nova coluna: classes = ['Baixa', 'Média', 'Alta'].

Aplicamos o método "cut" e apresentamos o novo histograma. 


```python
## Encontra os valores para separar os dados de potência em 3 classes
valores = np.linspace(min(frame_dados["potência(HP)"]), max(frame_dados["potência(HP)"]), 4)

## Definimos os nomes das classes de potência
classes=['Baixa', 'Média', 'Alta']

## Cria uma nova coluna no frame de dados separando as potências em 3 classes usando o método cut do Panda
frame_dados['classe-potência'] = pd.cut(frame_dados['potência(HP)'], valores, labels = classes, include_lowest = True )

## Plota o gráfico de barras para a nova coluna
pyplot.bar(classes, frame_dados["classe-potência"].value_counts())

## Define o rótulo dos eixos e do gráfico.
plt.pyplot.xlabel("Potência")
plt.pyplot.ylabel("Quantidade")
plt.pyplot.title("Classes de Potência")
```

Verifique no frame de dados que existe uma nova coluna de "classe-potência" na tabela e que seus valores estão distribuídos em 3 classes ("Baixa", "Média", "Alta") e não mais em 57 valores como na coluna "potência(HP)". Verifique a diferença em se mostrar os dados da nova coluna usando um gráfico de barras como acima e como um histograma, como foi feito inicialmente.

<h2 id="indicadoras">Variáveis indicadoras</h2>
<b>O que é uma variável indicadora?</b>
<p>
    Uma variável indicadora é uma variável numérica utilizada para rotular categoriasAn indicator variable (or dummy variable) is a numerical variable used to label categories. Note que neste tipo de variável os valores numéricos não  possuem um valor significativo, por este motivo é possível encontrar na literatura estas variáveis com o nome de "dummies variables". 
</p>

<b>Porque utilizar variáveis indicadoras?</b>
<p>
    Elas auxiliam na análise de regressão de variáveis categóricas. Este assunto será estudado mais a frente. 
</p>
<b>Exemplo</b>
<p>
    Considere a coluna "tipo de combustível" ela possui apenas dois valores, "gas" ou "diesel". O processo de regressão não é capaz de trabalhar com strings, apenas com números. Se quisermos utilizar este atributo em um processo de regressão teremos que converter a coluna "tipo de combustível" em variáveis indicadoras.
</p>

<p>
    Para isto iremos utilizar o método do Panda <code>.get_dummies</code> de forma a criar valores numéricos para os tipos diferentes de combustível. 
</p>

```python
## O código abaixo cria a variável indicadora a partir da coluna "tipo de combustível" e mostra o resultado
variavel_indicadora_1 = pd.get_dummies(frame_dados["tipo de combustível"])
variavel_indicadora_1.head()
```

<h2 id="agrupamento-basico">Agrupamento - Introdução</h2>

<p>O método <code>groupby</code> faz o agrupamento de dados por categorias. Os dados são agrupados baseados em uma ou mais variáveis e a análise é feita em grupos individuais. </p>

<p>Por exemplo, vamos agrupar a variável "rodas-tração". Note que existem 3 tipos diferentes de tração.</p>

<p>Se quisermos saber, na média, qual o tipo de tração mais valioso, podemos agrupar e, em seguida, determinar a média. </p>

<p>Podemos fazer este processo considerando ainda mais de uma variável. Todos estes processos estão apresentados nas linhas de código abaixo.  </p>


```python
## Determina as classes existentes na variável "rodas-tração"
frame_dados['rodas-tração'].unique()
```

```python
## Agrupa as 3 variáveis em um novo frame de dados
fd_grupo_1 = frame_dados[['rodas-tração','estilo-corpo','preco']]
```

```python
## Determina a média para cada tipo de "rodas-tração"
fd_grupo = fd_grupo_1.groupby(['rodas-tração'],as_index=False).mean()
fd_grupo
```

```python
## Determina a média para cada par de valores entre o produto cartesiano entre "rodas-tração" e "estilo-corpo"
teste_agrupamento = fd_grupo_1.groupby(['rodas-tração','estilo-corpo'],as_index=False).mean()
teste_agrupamento
```

<p>Estes dados agrupados são mais faceis de serem visualizados se forem apresentados em uma tabela pivotada como em Excel, onde uma variável é colocada em coluna e outra em linha. Podemos converter um frame de dados em uma tabela pivotada utilizando o metodo <code> pivot()</code> conforme apresentado abaixo. </p>

<p>Neste caso iremos colocar a variável "rodas-tração" na linha e a variável "estilo-corpo" como colunas da tabela: </p>

```python
tabela_pivot = teste_agrupamento.pivot(index='rodas-tração',columns='estilo-corpo')
tabela_pivot
```

<p>Em alguns casos, se não existir valor para uma posição na tabela pivotada o valor NaN será colocado. É possível substituir  este valor por 0 ou outro valor qualquer. O tópico de dados faltantes é bem complexo e pode ser tratado de várias formas. </p>

```python
## Coloca 0 nas posições onde não existe valor.
tabela_pivot = tabela_pivot.fillna(0) 
tabela_pivot
```

<h2 id="correlação">Correlação e Causalidade</h2>

<p><b>Correlação</b>: uma medida da interdependência entre variáveis.</p>

<p><b>Causalidade</b>: a relação entre causa e efeito entre duas variáveis.</p>

<p>É importante entender a diferença entre estas duas medidas e que a correlação não implica em causalidade. Determinar a correlação é bem mais simplesque causalidade. A determinação de causalidade requer experimentos independentes. </p>

<h3>Correlação de Pearson</h3>
<p>A correlação de Pearson mede a dependência linear entre duas variáveis X e Y. O resultado é um coeficiente que possui um valor entre -1 e 1, inclusive, onde: </p>
<ul>
    <li><b>1</b>: Correlação linear total e positiva.</li>
    <li><b>0</b>: Sem correlação linear, o mais provável é que uma variável não tenha efeito algum sobre a outra.</li>
    <li><b>-1</b>: Correlação linear total e negativa.</li>
</ul>

<p>A correlação de Pearson é o padrão utilizado no método <code> corr()</code>. Lembrando que a correlação é determinada apenas para variáveis do tipo float e int.  </p>

```python
frame_dados.corr()
```

Em alguns casos é possível saber a significância da estimativa da correlação. 

<b>Valor-P</b>: 
<p>O valor-P é uma probabilidade que indica a significância estatística entre as duas variáveis. Geralmente é escolhido um valor de significância da ordem de 0.05, que implica em uma confidência de 95% na correlação entre as variáveis. </p>

Por convenção temos que:
<ul>
    <li>valor-P é $<$ 0.001: existe uma forte evidência que a correlação é significativa.</li>
    <li>valor-P é $<$ 0.05: existe uma evidência moderada que a correlação é significativa.</li>
    <li>valor-P é  $<$ 0.1: existe uma evidência fraca que a correlação é significativa.</li>
    <li>valor-P é  $>$ 0.1: não existe evidência que a correlação seja significativa.</li>
</ul>

É possível calcular este valor utilizando o método <code>stats()</code> da biblioteca <b>scipy</b>.

Vamos determinar estes valores (correlação e valor-P) de diversos pares de variáveis e verificar a qualidade destas variáveis para prever o valor do preço do carro.

```python
from scipy import stats
```

```python
corr_pearson, valor_p = stats.pearsonr(frame_dados['base-rodas'], frame_dados['preco'])
print("O coeficiente de correlação de Pearson = ", corr_pearson, " com valor-P = ", valor_p)  
```

<h2>ANOVA: Análise de Variância</h2>
<p>A análise de variância (ANOVA) é um método estatístico utilizado para testar se existem diferenças significativas entre as médias de dois ou mais grupos de variáveis. ANOVA retorna dois parâmetros:</p>

<p><b>F-test score</b>: ANOVA assume que a média dos dois grupos é a mesma e determina quanto as médias reais divergem e relata isto através do F-test score. Quanto maior este valor maior a divergência entre as médias. </p>

<p><b>valor-P</b>: O valor de P mostra quão significativo, estatisticamente falando, é o valor do F-Test. </p>

<p>Se a variável preço for fortemente correlacionada com uma outra variável que estejamos analisando, espera-se um valor considerável de F-test score e um valor pequeno de valor-P. </p>

<h3>Rodas de Tração</h3>

<p>Como ANOVA analisa a diferença entre grupos diferentes da mesma variável o método <code>groupby()</code> é bastante útil. Como ANOVA faz a média dos dados automaticamente não é preciso determinar a média de antemão. </p>

<p>Vamos ver se o tipo de tração possui algum impacto no preço dos carros, vamos agrupar estas variáveis. </p>



```python
## Faz o agrupamento das variáveis "rodas-tração" e "preco"
teste_agrupamento_2=fd_grupo_1[['rodas-tração', 'preco']].groupby(['rodas-tração'])
teste_agrupamento_2.head(2)
```

```python
## Determina o grupo de carros "4wd" e seus preços
teste_agrupamento_2.get_group('4wd')['preco']
```

```python
## Aplica ANOVA entre os grupos de carros de diferentes trações e seus preços
f_val, p_val = stats.f_oneway(teste_agrupamento_2.get_group('fwd')['preco'], teste_agrupamento_2.get_group('rwd')['preco'], teste_agrupamento_2.get_group('4wd')['preco'])  
 
print( "ANOVA resultados: F=", f_val, ", P =", p_val)   
```

Este é um bom resultado, o valor grande de F mostra uma correlação forte e com o valor-P próximo a 0 implica que a significância estatística é quase certa. Mas este resultado entre as três variáveis testadas indica que elas são todas altamente correlacionadas?  

Vamos fazer a análise novamente, mas agora considerando o tipo de tração dois a dois.

<b>fwd e rwd</b>

```python
f_val, p_val = stats.f_oneway(teste_agrupamento_2.get_group('fwd')['preco'], teste_agrupamento_2.get_group('rwd')['preco'])  
 
print( "ANOVA resultados: F=", f_val, ", P =", p_val )
```

<b>4wd e rwd</b>

```python
f_val, p_val = stats.f_oneway(teste_agrupamento_2.get_group('4wd')['preco'], teste_agrupamento_2.get_group('rwd')['preco'])  
   
print( "ANOVA resultados: F=", f_val, ", P =", p_val)   
```

<b>4wd e fwd</b>

```python
f_val, p_val = stats.f_oneway(teste_agrupamento_2.get_group('4wd')['preco'], teste_agrupamento_2.get_group('fwd')['preco'])  
 
print("ANOVA resultados: F=", f_val, ", P =", p_val)   
```

Veja que a correlação entre "fwd" e "rwd" é ainda mais forte e com uma valor de P bem próximo de 0. Este resultado já não é tão bom para os casos de comparação com os "4wd". Isto, provavelmente, ocorre pois o número de veículos "4wd" é pequeno na base de dados.
